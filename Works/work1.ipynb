```python
df = pd.read_csv(
    'https://raw.githubusercontent.com/Dankrou/Automated_Data_Processing_and_Control_Systems/master/CSV/olympics.csv', #использую свой личный гитхаб для хранения таблицы
    index_col = 0, # Номер столбца который будет индексом
    skiprows = 0 # Номер сроки которая будет заголовком
)

names_ids = df.index.str.split(' ') 

df.index = names_ids.str[0] 



df = df.drop('Totals')
df.head(15) 
```

```python
df.iloc[1].name
```

```python
df.iloc[1].loc[['Gold', 'Silver', 'Bronze']]
```

```python
cost = max(df['Gold.1']) 

answer = df[(df['Gold.1'] == cost)]

print('Страна -', answer.index[0])

answer
```

```python
diff = df['Silver'] - df['Silver.1'] # Разница между летними и зимними серебряными медалями

max_diff = max(diff) # Наибольная максимальная разница между количеством летних и зимних серебряных медалей

answer = df[(diff == max_diff)]

print('Страна - {} [{}]'.format(answer.index[0], max_diff))

answer
```
